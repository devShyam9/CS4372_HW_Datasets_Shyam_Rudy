```python
import pandas as pd
```

```python
# Read the file
url = "https://raw.githubusercontent.com/devShyam9/CS4372_HW_Datasets_Shyam_Rudy/refs/heads/main/Tetuan%20City%20power%20consumption.csv"

df = pd.read_csv(url)
```

```python
# Convert DateTime to datetime type
df['DateTime'] = pd.to_datetime(df['DateTime'])
print(df.shape)
df.head()
```

```python
# Examining the attributes and distributions
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Summary stats
df.describe()
```

```python
# Plot histograms with KDE for each numeric column to check distributions
numeric_cols = df.select_dtypes('float64').columns
for col in numeric_cols:
  sns.histplot(df[col], kde=True)
  plt.title(col)
  plt.show()
```

```python
# Feature engineering: extract time-based features from DateTime
df['Hour'] = df['DateTime'].dt.hour
df['Day'] = df['DateTime'].dt.day
df['Month'] = df['DateTime'].dt.month
df['DayOfWeek'] = df['DateTime'].dt.dayofweek
df.head()
```

```python

from sklearn.preprocessing import StandardScaler
```

```python
# Define features and target for modeling
features = ['Temperature', 'Humidity', 'Wind Speed', 'general diffuse flows', 'diffuse flows', 'Hour', 'DayOfWeek']
target = 'Zone 1 Power Consumption'

X = df[features]
y = df[target]

scaler = StandardScaler()
X_Scaled = scaler.fit_transform(X)
```

```python
# Correlation analysis between features and the target
corr = df[features + [target]].corr()
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
# Split the data into training and testing sets (80/20 split)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_Scaled, y, test_size=0.2, random_state=42)
print(X_train.shape, X_test.shape)
```

```python
# Show correlations of all features with the target, sorted by strength
corr_target = corr['Zone 1 Power Consumption'].drop('Zone 1 Power Consumption')
print(corr_target.sort_values(ascending=False))
```

```python
# Model construction: define final features and target
final_features = ['Hour', 'Temperature', 'Humidity', 'Wind Speed', 'general diffuse flows']
target = 'Zone 1 Power Consumption'
X = df[final_features]
y = df[target]
# Scale features before training
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_Scaled = scaler.fit_transform(X)
# Split into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_Scaled, y, test_size=0.2, random_state=42)
```

```python
X_train.shape, X_test.shape
```

```python
# Hyperparameter tuning for SGDRegressor
from itertools import product
import pandas as pd
from sklearn.linear_model import SGDRegressor
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.model_selection import train_test_split

# Create a small validation set from the training data
X_tr, X_val, y_tr, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=42)
# Define parameter grid to test
param_grid = {
    "loss": ["squared_error"],
    "penalty": ["l2", "elasticnet"],
    "alpha": [1e-5, 1e-4, 1e-3],
    "learning_rate": ["adaptive"],
    "eta0": [0.01],
    "l1_ratio": [0.15, 0.5],  # only applies if penalty='elasticnet'
}
# Helper function to loop through all parameter combinations
def grid_iter(grid):
    keys = list(grid.keys())
    for vals in product(*grid.values()):
        yield dict(zip(keys, vals))

records = []
# Train and evaluate SGDRegressor for each parameter set
for params in grid_iter(param_grid):
    model = SGDRegressor(
        max_iter=5000,
        tol=1e-4,
        early_stopping=True,
        validation_fraction=0.1,
        n_iter_no_change=5,
        random_state=42,
        **params
    )
    model.fit(X_tr, y_tr)
# Predictions on training and validation sets
    y_tr_pred = model.predict(X_tr)
    y_val_pred = model.predict(X_val)
# Record metrics for this parameter configuration
    records.append({
        **params,
        "train_R2": r2_score(y_tr, y_tr_pred),
        "val_R2": r2_score(y_val, y_val_pred),
        "train_MSE": mean_squared_error(y_tr, y_tr_pred),
        "val_MSE": mean_squared_error(y_val, y_val_pred),
    })
# Create a DataFrame of results, sorted by validation R^2
tuning_df = pd.DataFrame(records).sort_values("val_R2", ascending=False).reset_index(drop=True)
tuning_df.head(12)


```

```python
# OLS regression
import statsmodels.api as sm
# Add intercept term to the scaled feature matrices
X_train_ols = sm.add_constant(X_train)
X_test_ols = sm.add_constant(X_test)

# Fit OLS on the training data
ols_model = sm.OLS(y_train, X_train_ols).fit()

# Predict on train and test
y_test_pred_ols = ols_model.predict(X_test_ols)
y_train_pred_ols = ols_model.predict(X_train_ols)
# Report OLS metrics
print("Training MSE:", mean_squared_error(y_train, y_train_pred_ols))
print("Training R^2:", r2_score(y_train, y_train_pred_ols))
print("Testing MSE:", mean_squared_error(y_test, y_test_pred_ols))
print("Testing R^2:", r2_score(y_test, y_test_pred_ols))

# Show full OLS summary
ols_model.summary()

```

```python
# Tuned SGDRegressor
best_params = {
    "loss": "squared_error",
    "penalty": "l2",
    "alpha": 1e-4,
    "learning_rate": "adaptive",
    "eta0": 0.01,
}

from sklearn.linear_model import SGDRegressor
from sklearn.metrics import r2_score, mean_squared_error
# Fit SGD with best parameters on the training data
final_sgd = SGDRegressor(max_iter=5000, tol=1e-4, random_state=42, **best_params)
final_sgd.fit(X_train, y_train)

print("SGD Train R²:", r2_score(y_train, final_sgd.predict(X_train)))
print("SGD Test  R²:", r2_score(y_test,  final_sgd.predict(X_test)))
print("SGD Train MSE:", mean_squared_error(y_train, final_sgd.predict(X_train)))
print("SGD Test  MSE:", mean_squared_error(y_test,  final_sgd.predict(X_test)))
```
